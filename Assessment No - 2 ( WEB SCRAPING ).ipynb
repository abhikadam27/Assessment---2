# 1. header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
url = 'https://en.wikipedia.org/wiki/Main_Page'
def wikipedia_headers(url):
   
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    titles = soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
    print('List all the header tags :', *titles, sep='\n\n')
    return

wikipedia_headers(url)
```

# 2.IMDB’s Top rated 100 movies’ data

```python
url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
def IMDB_Movie(url):    
    

    page = requests.get(url)

    soup = BeautifulSoup(page.text, "html.parser")

    titles = soup.find_all("td",{"class":"titleColumn"})[0:100]

    ratings = soup.find_all("td",class_="ratingColumn imdbRating")[0:100]

    movie_rating = []
    for i in ratings:
        movie_rating.append(i.text.replace('\n',''))

    year =soup.find_all('span',class_="secondaryInfo")[0:100]

    movie_name=[]
    movie_year=[]
    for i in titles:
        movie_name.append(i.a.text)
    for i in year:
        movie_year.append(i.text)

    data = list(zip(movie_name,movie_year,movie_rating))

    import pandas as pd
    df = pd.DataFrame(data,columns =["Movie Name","Year of release","Rating"])

    df
    return(df)
IMDB_Movie(url)
```

# 4. books data from ‘www.bookpage.com’

```python
url = "https://bookpage.com/reviews"
def BookPage(url):
    page = requests.get(url)

    soup = BeautifulSoup(page.text,"html.parser")

    book_name = []
    for i in soup.find_all("h4",class_="italic"):
        book_name.append(i.text.replace('\n',''))

    auther_name = []
    for i in soup.find_all("p",class_="sans bold"):
        auther_name.append(i.text.replace('\n',''))

    genre = []
    for i in soup.find_all("p",class_="genre-links hidden-phone"):
        genre.append(i.text.replace('\n',''))

    review = []
    for x in soup.find_all("div",class_="read-full"):
        review.append('https://bookpage.com/'+str(x.a["href"]))

    data = list(zip(book_name,auther_name,genre,review))

    import pandas as pd
    book_details = pd.DataFrame(data, columns = ["Book Name","Auther Name","Genre","Review"])
    return(book_details.head(5))
BookPage(url)
```

# 5.1 ICC Top 10 ODI teams in men’s cricket

```python
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
def Top_Mens_Teams(url):
    page = requests.get(url)

    soup = BeautifulSoup(page.text,"html.parser")

    table = soup.find("tbody")

    team_name = table.find_all("span",class_="u-hide-phablet")[0:10]

    teams = []
    for i in team_name:
        teams.append(i.text)

    match =  table.find_all("td",class_="table-body__cell u-center-text")

    f_match = table.find("td",class_="rankings-block__banner--matches").text

    f_points = table.find("td",class_="rankings-block__banner--points").text

    matches = [f_match,f_points]
    for i in match:
        matches.append(i.text)

    m_records = matches[0:20:2]

    m_points = matches[1:20:2]

    f_rating = table.find("td",class_="rankings-block__banner--rating u-text-right").text.replace('\n','').strip()

    ratings = table.find_all("td",class_="table-body__cell u-text-right rating")[0:9]

    team_rating = [f_rating]
    for i in ratings:
        team_rating.append(i.text)

    data = list(zip(teams,m_records,m_points,team_rating))

    import pandas as pd
    df = pd.DataFrame(data, columns = ["Team Name","Matches","Points","Ratings"])
    return(df)
Top_Mens_Teams(url)
```

# 5.2 Top 10 ODI Batsmen

```python
url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"
def Top_Batsmen(url):    
    page = requests.get(url)

    soup = BeautifulSoup(page.text,"html.parser")

    table = soup.find("table",class_="table rankings-table")

    first_player = table.find("div",class_="rankings-block__banner--name-large").text

    first_team = table.find("div",class_="rankings-block__banner--nationality").text.replace('\n','')

    first_rating = table.find("div",class_="rankings-block__banner--rating").text

    players = table.find_all("td",class_="table-body__cell rankings-table__name name")[0:9]

    player = [first_player]
    for i in players:
        player.append(i.text.replace('\n',''))

    teams= table.find_all("span",class_="table-body__logo-text")[0:9]

    team = [first_team]
    for i in teams:
        team.append(i.text)

    ratings = table.find_all("td",class_="table-body__cell rating")[0:9]

    rating = [first_rating]
    for i in ratings:
        rating.append(i.text)

    data = list(zip(player,team,rating))

    import pandas as pd
    df = pd.DataFrame(data, columns = ["Player Name","Team","Rating"])
    return(df)
Top_Batsmen(url)
```

# 7. Mobile phones under Rs. 20,000 listed on Amazon.in.

```python
def Amazon_mobiles(url):
    url = "https://www.amazon.in/s?k=all+mobiles+under+20000rs&ref=nb_sb_noss"
    HEADERS = ({'User-Agent':'Mozilla/5.0 (X11; Linux x86_64)AppleWebKit/537.36 (KHTML, like Gecko)Chrome/44.0.2403.157 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})

    Page = requests.get(url, headers=HEADERS)

    soup = BeautifulSoup(Page.text,"html.parser")

    Names = []
    price = []
    A_rating = []

    for i in soup.find_all('a', class_='a-link-normal a-text-normal'):
        string = i.text
        Names.append( string.strip() )
    for i in soup.find_all("span",class_="a-price-whole"):
        price.append('Rs '+str(i.text))
    for i in soup.find_all("span",class_="a-icon-alt"):
        A_rating.append(i.text)

    image = []
    for i in soup.find_all("img"):
        image.append(i["src"])

    data = list(zip(Names,price,image,A_rating))

    import pandas as pd
    df = pd.DataFrame(data, columns = ["Mobile name","Price","Image Link","Average Rating"])
    df
    return(df)
Amazon_mobiles(url)
```

# 8.  Information about the local weather from the National Weather Service
website of USA, https://www.weather.gov/ for the city, San Francisco.

```python
url = "http://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.416"
def weather_forcast(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    seven_day = soup.find(id="seven-day-forecast")
    forecast_items = seven_day.find_all(class_="tombstone-container")
    tonight = forecast_items[0]

    period = tonight.find(class_="period-name").get_text()
    short_desc = tonight.find(class_="short-desc").get_text()
    temp = tonight.find(class_="temp").get_text()

    img = tonight.find("img")
    desc = img['title']

    period_tags = seven_day.select(".tombstone-container .period-name")
    periods = [pt.get_text() for pt in period_tags]

    short_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
    temps = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
    descs = [d["title"] for d in seven_day.select(".tombstone-container img")]

    import pandas as pd
    weather = pd.DataFrame({
            "period": periods, 
            "short_desc": short_descs, 
            "temp": temps, 
            "description":descs
        })
    weather
    return(weather)
weather_forcast(url)
```

# 9. Fresher job listings from ‘https://internshala.com/’.

```python
url = "https://internshala.com/fresher-jobs"
def Fresher_Jobs(url):
    page = requests.get(url)

    soup = BeautifulSoup(page.text,"html.parser")

    job_title = soup.find_all("div",class_="heading_4_5 profile")

    job_titles = []
    for i in job_title:
        job_titles.append(i.text.replace('\n',''))

    company_name = soup.find_all("div",class_="heading_6 company_name")

    company_names = []
    for i in company_name:
        company_names.append(i.text.replace('\n','').strip())

    ctc = soup.find_all("div",class_="item_body")

    CTC = []
    for x in ctc:
        CTC.append(x.text.replace('\n','').strip())

    CTC_ =CTC[1:130:3]
    apply_date = CTC[2:130:3]

    data = list(zip(job_titles,company_names,CTC_,apply_date))

    import pandas as pd
    df = pd.DataFrame(data, columns = ["Job Title","Company Name","CTC","Apply Date"])
    return(df)
Fresher_Jobs(url)
```
